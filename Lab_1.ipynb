```python
#Libraries
import pandas as pd
import sklearn
import sklearn.datasets
import sklearn.neighbors
import numpy as np
```

```python
#Load Data
df=pd.read_csv("/content/drive/MyDrive/dataset/heart.csv")
df.head()
```

```python
#Convert Categorical Values into Numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ['sex', 'dataset', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'thal']
one_hot = OneHotEncoder()
transformer = ColumnTransformer([('one_hot',
                                  one_hot,
                                  categorical_features)],
                                remainder='passthrough')
transformed_x = transformer.fit_transform(df)
transformed_x
```

```python
#Encoded the Different Parameters of each Categorical Features into Numbers
pd.DataFrame(transformed_x)
```

```python
#Removed All NaN (missing rows)
df = df.dropna()
print(df)
```

```python
#Splitting Data into Features (x) and Labels (y)
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
print(transformed_x)
print(y)
```

```python
#Splitting Data into 60% Training, 20% Development, and 20% Testing
from sklearn.model_selection import train_test_split
x_train, x_temp, y_train, y_temp = train_test_split(x, y, test_size=0.4, random_state=42)
x_dev, x_test, y_dev, y_test = train_test_split(x_temp, y_temp, test_size=0.5, random_state=42)
print(x_train.shape)
print(x_dev.shape)
print(x_test.shape)
```

```python
#Classification Models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
```

```python
#Performance Metrics
results = {'model': [], 'precision': [], 'recall': [], 'f1_score': [], 'accuracy': []}
```

```python
#Hyperparameters for Train/Dev
#Logistic Regression Hyperparameters
log_reg_params = [{'C': 0.01, 'solver': 'liblinear'}, {'C': 0.1, 'solver': 'liblinear'}, {'C': 1.0, 'solver': 'liblinear'}, {'C': 10, 'solver': 'liblinear'}, {'C': 100, 'solver': 'liblinear'}]

#Random Forest Hyperparameters
rf_params = [{'n_estimators': 1000, 'max_depth': 50}, {'n_estimators': 500, 'max_depth': 30}, {'n_estimators': 100, 'max_depth': 5}, {'n_estimators': 200, 'max_depth': 7}, {'n_estimators': 300, 'max_depth': 10}]

#SVM Hyperparameters
svm_params = [{'C': 0.1, 'kernel': 'linear'}, {'C': 0.5, 'kernel': 'sigmoid'}, {'C': 2, 'kernel': 'rbf', 'gamma': 'auto'}, {'C': 1, 'kernel': 'rbf', 'gamma': 'scale'}, {'C': 10, 'kernel': 'poly', 'degree': 2}]
```

```python
#Evaluate Logistic Regression - Train/Dev
def evaluate_log_reg(log_reg, logistic_regression, transformed_x_train, y_train, transformed_x_dev, y_dev, params, results):
    log_reg.fit(transformed_x_train, y_train)
    y_pred = log_reg.predict(transformed_x_dev)
    precision = precision_score(y_dev, y_pred, average='weighted', zero_division=1)
    recall = recall_score(y_dev, y_pred, average='weighted', zero_division=1)
    f1 = f1_score(y_dev, y_pred, average='weighted', zero_division=1)
    accuracy = accuracy_score(y_dev, y_pred)
    results['logistic_regression'].append(f'[log_reg][params]')
    results['precision'].append(precision)
    results['recall'].append(recall)
    results['f1_score'].append(f1)
    results['accuracy'].append(accuracy)
```

```python
results = {'logistic_regression': [], 'precision': [], 'recall': [], 'f1_score': [], 'accuracy': []}
```

```python
encoder = OneHotEncoder(handle_unknown='ignore')

#Fit the encoder on the training data
encoder.fit(x_train)

#Transform both training and development data
x_train = encoder.transform(x_train)
x_dev = encoder.transform(x_dev)

#Proceed with training and evaluation
for params in log_reg_params:
    log_reg = LogisticRegression(**params)
    evaluate_log_reg(log_reg, "LogisticRegression", x_train, y_train, x_dev, y_dev, params, results)
```

```python
#Results for Logistic Regression - Train/Dev
results_df = pd.DataFrame(results)
print(results_df)
```

```python
#Best Precision Configuration for Logistic Regression - Train/Dev
best_precision = results_df.sort_values(by='precision', ascending=False).iloc[0]
print("\nBest Configuration Based on precision:")
print(best_precision)
```

```python
#Best F1 Score for Logistic Regression - Train/Dev
best_f1_score = results_df.sort_values(by='f1_score', ascending=False).iloc[0]
print("\nBest Configuration Based on F1-Score:")
print(best_f1_score)
```

```python
#Best Recall for Logistic Regression - Train/Dev
best_recall = results_df.sort_values(by='recall', ascending=False).iloc[0]
print("\nBest Configuration Based on recall:")
print(best_recall)
```

```python
#Best Accuracy for Logistic Regression - Train/Dev
best_accuracy = results_df.sort_values(by='accuracy', ascending=False).iloc[0]
print("\nBest Configuration Based on accuracy:")
print(best_accuracy)
```
